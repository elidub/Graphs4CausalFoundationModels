```python
# Fetch and inspect a Weights & Biases run history
# - Configure your entity, project, and run ID below.
# - The history may be large; we sample or filter columns for display.

import os
import pandas as pd
import wandb

# Configure your W&B identifiers
ENTITY = os.environ.get("WANDB_ENTITY", "arik_reuter-max-planck-society")
PROJECT = os.environ.get("WANDB_PROJECT", "DoPFN_v2")
RUN_ID = os.environ.get("WANDB_RUN_ID", "kfjsjd0c")  # override via env if needed

api = wandb.Api()
run_path = f"{ENTITY}/{PROJECT}/{RUN_ID}"
print(f"Loading run: {run_path}")
run = api.run(run_path)

# Fetch run history – this can be large; by default W&B paginates efficiently.
hist = run.history(samples=100000)  # increase/decrease as needed
print(f"History rows fetched: {len(hist)}")

# Optionally select a subset of interesting columns for quick view
interesting_prefixes = [
    "train/",
    "eval/syn_",
    "real/",
    "checkpoints/",
    "summary/",
]

cols = [c for c in hist.columns if any(c.startswith(p) for p in interesting_prefixes)]
cols = sorted(set(cols + ["_step", "_runtime", "_timestamp"]))

preview = hist[cols] if cols else hist
print("Preview columns:", list(preview.columns)[:20])
preview.tail(10)
```
